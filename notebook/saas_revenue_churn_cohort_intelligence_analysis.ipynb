### 📝 Handling Active Subscriptions and Zero-MRR Records

Several subscription records contain a **missing `end_date`** and/or **zero recurring revenue (`MRR = 0`)**.  
These observations are intentionally retained because they can represent commercially valid states rather than erroneous records.

In a SaaS environment:

- A missing `end_date` often indicates an **ongoing active subscription**
- Zero MRR can legitimately reflect **free plans, trial periods, 100% discounts, promotional access, or delayed billing activation**

Removing these records would artificially shrink the active customer base and distort:

- Retention rates  
- Churn calculations  
- Active subscription counts  
- Revenue interpretation  

To preserve analytical accuracy, the workflow creates a **business-friendly subscription status classification** that separates:

- **Active paid subscriptions**
- **Active free / zero-MRR subscriptions**
- **Ended or churned subscriptions**

This treatment keeps the analysis faithful to how subscription businesses actually operate.

## Creating Subscription Status

```python
subscriptions_df["subscription_status"] = subscriptions_df.apply(
    lambda row: "Active - Paid"
    if pd.isna(row["end_date"]) and row["mrr_amount"] > 0
    else "Active - Free"
    if pd.isna(row["end_date"]) and row["mrr_amount"] == 0
    else "Ended",
    axis=1
)
```

### 💬 Handling Missing Satisfaction Feedback

Some support tickets do not include a customer satisfaction score.  
These cases are retained because **absence of feedback is not equivalent to poor feedback** and should not be force-filled with arbitrary values.

To avoid introducing bias:

- Satisfaction scores remain **null** where no response was provided
- A separate indicator identifies whether **feedback was submitted or not**
- Support quality analysis can therefore distinguish between:
  - **Measured satisfaction**
  - **No recorded feedback**

```python
support_tickets_df["feedback_status"] = support_tickets_df["satisfaction_score"].apply(
    lambda x: "Feedback Provided" if pd.notna(x) else "Feedback Not Provided"
)
```

## Q.1::What is the distribution of accounts across different plan tiers?

```python
#  Distribution of accounts across different plan tiers

plan_counts = accounts_df["plan_tier"].value_counts()
print(plan_counts)
```

```python
plt.figure(figsize=(8, 5))
plt.bar(plan_counts.index, plan_counts.values, color="darkblue")
plt.title("Distribution of Accounts Across Plan Tiers")
plt.xlabel("Plan Tier")
plt.ylabel("Number of Accounts")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

## Q.2::How are account signups trending over time (monthly growth)?


```python
monthly_signups = (
    accounts_df
    .set_index("signup_date")
    .resample("M")
    .size()
    .reset_index(name="signup_count")
)
print(monthly_signups)
```

```python
sns.lineplot(
    data=monthly_signups,
    x="signup_date",
    y="signup_count",
    color="pink",
    marker="o"
)

plt.title("Monthly Account Signup Trend")
plt.xlabel("Signup Month")
plt.ylabel("Number of Signups")
plt.tight_layout()
plt.show()
```

## Q.3::What proportion of accounts started as trial vs non-trial?

```python
trial_counts = accounts_df["is_trial"].value_counts()
print(trial_counts)
```

```python
plt.figure(figsize=(6, 6))
plt.pie(
    trial_counts.values,
    labels=["Non-Trial", "Trial"],
    autopct="%1.1f%%",
    colors=["darkblue", "pink"],
    startangle=90
)
plt.title("Proportion of Trial vs Non-Trial Accounts")
plt.tight_layout()
plt.show()
```

## Q.4::How is monthly revenue (MRR) distributed across subscriptions?

```python
sns.histplot(
    data=subscriptions_df,
    x="mrr_amount",
    bins=30,
    color="darkblue",
    kde=True
)

plt.title("Distribution of Monthly Recurring Revenue (MRR)")
plt.xlabel("Monthly Revenue (MRR)")
plt.ylabel("Number of Subscriptions")
plt.tight_layout()
plt.show()
```

## Q.5::Which features are used most frequently by users?


```python
feature_usage = (
    feature_usage_df
    .groupby("feature_name")["usage_count"]
    .sum()
    .sort_values(ascending=False)
)
print (feature_usage)
```

```python
plt.figure(figsize=(9, 5))
plt.bar(feature_usage.index, feature_usage.values, color="darkblue")
plt.title("Most Frequently Used Features")
plt.xlabel("Feature Name")
plt.ylabel("Total Usage Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

## Q.6::How does feature usage intensity vary across different features?

```python
feature_intensity = (
    feature_usage_df
    .groupby("feature_name")["usage_count"]
    .sum()
    .reset_index()
    .sort_values(by="usage_count", ascending=False)
)
print(feature_intensity)
```

```python
sns.barplot(
    data=feature_intensity,
    x="feature_name",
    y="usage_count",
    color="pink"
)

plt.title("Feature Usage Intensity Across Features")
plt.xlabel("Feature Name")
plt.ylabel("Total Usage Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
